```python
# analyza kmeru
import pandas as pd
import os
pathh = "/home/mg/ChimeraProject/pickles/"
filenames = os.listdir(pathh)
infonames = list(filter(lambda x: ".info" in x,filenames))
infonames
above_thres, uniq_per_fam = [],[]
for i in infonames:
    df_kmer_info = pd.read_csv(f"/home/mg/ChimeraProject/pickles/{i}",sep="\t")
    
    above_thres.append(df_kmer_info.iloc[[1],[0]].values)
    uniq_per_fam.append(df_kmer_info.iloc[[-1],[0]].values)

above_thres = list(map(lambda x: int(x),above_thres))
uniq_per_fam = list(map(lambda x: int(x),uniq_per_fam))
```

```python
above_thres = above_thres[:15]
uniq_per_fam = uniq_per_fam[:15]
```

```python
import numpy as np
import matplotlib.pyplot as plt
from math import log
categories = list(map(lambda x: x.replace("kmers_unite_","").replace(".info",""),infonames))
cat_dic = {k:(v,w) for k,v,w in zip(categories,above_thres,uniq_per_fam)}
sorted_cat_dic = dict(sorted(cat_dic.items(), key=lambda x: (x[0].split("t")[0], log(int(x[0].split("t")[1])))))  
print(sorted_cat_dic)
#categories = sorted(categories,key=(lambda x:(x.split("t")[0],log(int(x.split("t")[1])))))
#print(categories)

categories = list(sorted_cat_dic.keys())[30:60]
values = [i[0] for i in sorted_cat_dic.values()][30:60]
group_size = 15
x_positions = []
gap = 1  # Adjust this to increase or decrease space between groups
for i in range(0, len(categories), group_size):
    x_positions.extend(np.arange(group_size) + (i + gap * (i // group_size)))
plt.figure(figsize=(20, 10))  
plt.bar(x_positions, values, color=['blue',"red"]*7+["blue"]+['blue',"red"]*7)
plt.xticks(x_positions, categories)
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Grouped Bar Plot with Spacing")
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()
```
